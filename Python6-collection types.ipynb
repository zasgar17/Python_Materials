####Lists — advanced lesson

Indexing, negative indexing, slicing

```python
a = [10, 20, 30, 40, 50]
print(a[0])    # 10
print(a[-1])   # 50
print(a[1:4])  # 20,30,40
print(a[:3])   # first 3
print(a[3:])   # from index 3 to end

#Slicing creates a new list.
```

Copy vs reference

```python
a = [1, 2, 3]
b = a
b[0] = 99
print(a)  # [99, 2, 3]
```

b = a does NOT make a new list

both names point to the same list in memory

```python
#Correct ways to copy

a = [1,2,3]
b = a[:]        # slice copy
c = a.copy()    # method copy
c[0]=99
d = list(a)     # constructor copy
print(c)
print(a)
print(d)
print(b)
```

“If I change b, will a change?”
Yes, if b=a. No, if b=a.copy().

###Sets (unique elements, fast membership)

```python
s = {1, 2, 3, 3, 3}
print(s)  # {1,2,3}

#Main concept:
#no duplicates
```

Add/remove and membership

```python
s.add(10)
s.remove(2)     # error if not exist
s.discard(99)   # safe, no error
print(3 in s)   # True/False
```

“why sets matter”:

Checking if something exists in a list can be slow for big lists

set membership is very fast (no need to explain hashing deeply)

**Convert list → set → list (remove duplicates)**

```python
nums = [1,2,2,3,3,3,4]
unique = list(set(nums))
print(unique)

#order may change; sets don’t guarantee order
```

Set operations

```python
A = {1,2,3,4}
B = {3,4,5}

print(A | B)  # union
print(A & B)  # intersection
print(A - B)  # difference
print(A ^ B)  # symmetric difference
```
