##Repititon Operators

####if-else recap

```python
x = 10
if x < 5:
    print("Big number")
else:
    print("Small number")
```

####for loops
A loop means: do something many times without writing it again.

```python
for i in range(5):
    print(i)
```

range(5) → numbers from 0 to 4

i → changing value each time

```python
for i in range(1, 6):
    print(i)
```

```python
for i in range(0, 100, 10):
    print(i)
```

####while loop

```python
x = 1
while x <= 5:

  print(x)
  x += 1
```

for → when you know how many times

while → when you don’t know when to stop

Important warning:

Forgetting x += 1 → infinite loop

**Loop + if together (very important!)**

This is where real logic starts.

```python
for i in range(1, 11):
    if i % 2 == 0:
        print(i, "is even")
    else:
        print(i, "is odd")
```

i % 2 == 0 → divisible by 2

####break and continue

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

**Example 1: Password check**

```python
password = "python123"

while True:
    user = input("Enter password: ")
    if user == password:
        print("Access granted")
        break
    else:
        print("Try again")
```

**Example 2: Sum of numbers**

```python
total = 0
for i in range(1, 6):
    total += i

print("Sum is:", total)
```
