Situation: grading

```python
grade=int(input())
if grade>=90:
  if grade>=93:
    print("A")
  else:
    print("A-")
elif grade>80:
  if grade>=86:
    print("B+")
  elif grade>=83:
    print("B")
  else:
    print("B-")
else:
  print("pass")
```
