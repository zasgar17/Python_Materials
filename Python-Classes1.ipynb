###BASIC CLASS & OBJECT

```python
class Student:
    pass

s1 = Student()
s2 = Student()

# Student → class
# s1, s2 → objects (instances)
```

###ATTRIBUTES (Variables inside objects)

```python
class Student:
    pass

s1 = Student()
s1.name = "Ali"
s1.age = 20

print(s1.name)

#objects can store data
```

###init METHOD (Constructor)

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("Ali", 20)
```

Instead of manually assigning:

 __init__ runs automatically

self = the current object

###METHODS (Functions inside classes)

```python
class Student:
    def __init__(self, name):
        self.name = name

    def say_hello(self):
        print("Hello, my name is", self.name)

s1 = Student("Ali")
s1.say_hello()
```

*Important Concept:*

self is ALWAYS the first parameter

###MODIFYING OBJECT STATE

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

acc = BankAccount(100)
acc.deposit(50)

print(acc.balance)  # 150
```

###MID-LEVEL EXAMPLES

Example 1: Rectangle

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

r = Rectangle(5, 3)
print(r.area())
```

Example 2: Counter

```python
class Counter:
    def __init__(self):
        self.count = 0

    def increase(self):
        self.count += 1

    def reset(self):
        self.count = 0
```

Example 3: Simple User System

```python
class User:
    def __init__(self, username, password):
        self.username = username
        self.password = password

    def check_password(self, pwd):
        return self.password == pwd
```
