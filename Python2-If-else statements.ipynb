###Common Beginner Mistakes (IMPORTANT)

Using "=" instead of "=="
Explain:

= → assign value

== → compare values

“One equal stores, two equals compare.”

```python
if age = 18:   #Wrong
if age == 18:  #True
```

Forgetting ":"

```python
if age > 18   #F
if age > 18:  #T
```

##Adding else (second path)

```python
age = int(input("Enter your age: "))

if age >= 18:
    print("You can vote")
else:
    print("You cannot vote")
```

###Multiple Conditions → elif

```python
score = float(input("Enter score: "))

if score >= 90:
    print("Excellent")
elif score >= 50:
    print("Pass")
else:
    print("Fail")
```

Python checks if

If False → checks elif

If all False → goes to else

Important rule:

“Python stops at the first True condition.”

```python
score = 95

if score >= 50:
    print("Pass")
elif score >= 90:
    print("Excellent")

    #this example is wrong, why?
```

Explanation:
Python never reaches elif

Order must be from strongest to weakest

```python
if score >= 90:
    print("Excellent")
elif score >= 50:
    print("Pass")

    #this is true, because we egin with strongest condition
```

###Logical Operators (and, or, not)

```python
age = 10
has_id = False

if age >= 18 and has_id:  # both must be true
    print("You may enter")
else:
    print("Entry denied")
```

```python
if age < 12 or age > 60: #or → one is enough
    print("Discount")
```

```python
is_raining = False

if not is_raining:  #not → opposite
    print("Go outside")
```

Nested if (advanced but useful):

“Sometimes we check a condition inside another condition.”

```python
age = 65

if age >= 18:
    if age >= 65:
        print("Senior")
    else:
        print("Adult")
else:
    print("Minor")
```

Example 1: Login System

```python
username = input("Username: ")
password = input("Password: ")

if username == "admin" and password == "1234":
    print("Login successful")
else:
    print("Wrong credentials")
```

Example 2: Weather Advice

```python
temp = int(input("Temperature: "))

if temp > 30:
    print("Hot")
elif temp > 15:
    print("Warm")
else:
    print("Cold")
```
