###Local vs Global Variables

```python
x = 10   # global

def test():
    x = 5  # local
    print(x)

test()
print(x)
```

Function has its own scope

###Functions Working With Lists

```python
def find_max(lst):
    max_value = lst[0]
    for num in lst:
        if num > max_value:
            max_value = num
    return max_value

nums=[45, 87, 1, 92, 45]
find_max(nums)
```

We pass list as argument

Function processes it

Returns result

###Writing Clean Programs (Mini Project)

```python
def get_numbers():
    return [1, 2, 3, 4]

def calculate_sum(numbers):
    return sum(numbers)

def main():
    nums = get_numbers()
    result = calculate_sum(nums)
    print(result)

main()
```

Big programs are divided into small functions

main() controls everything
