###Basic List Operations

####Adding

```python
names.append("Nigar")  #add item to end
names
```

```python
numbers = [1, 2, 3, 4]
numbers.insert(3, 100)    #add at position
numbers
```

```python
a = [1, 2]
b = [3, 4]
a.extend(b)            #add another list

a
```

append([3,4]) → nested list

extend([3,4]) → flat list

####Removing

```python
names = ["Aylin", "Sara", "Lala"]
names.remove("Sara")   #remove item
names
```

```python
nums = [1, 2, 3]
nums.remove(2)         #remove by value (it removes value, not place!)
nums
```

```python
nums = [1, 2, 3]
nums.pop(1)   # removes by index (index 1 here)
nums
```

```python
nums = [1, 2, 3]
nums.pop()    # removes last element
nums
```

```python
nums.clear()  #removes everyting
nums
```

####Finding elements

```python
numbers = [10, 20, 30, 40]
print(numbers[1])      #access one element
```

```python
#negative indexing
print(numbers[-1])  # last element → 40
print(numbers[-2])  # 30
```

####slicing (part of list)

```python
#start index included
#end index excluded

numbers = [1, 2, 3, 4, 5]

print(numbers[1:4])   # [2, 3, 4]
print(numbers[:3])
print(numbers[2:])
print(numbers[::-1])
```

####Searcing & checking

```python
names = ["Aylin", "Sara"]

print("Sara" in names)    # True
print("Nigar" in names)   # False
```

```python
names.index("Sara")  #finds position
```

```python
nums = [1, 2, 3, 3]
nums.count(3)   #counts repetition
```

####Ordering lists

```python
nums = [3, 1, 4, 2]
nums.sort()   #ascending (artan) order
nums
```

```python
#descending (azalan) order
nums.sort(reverse=True)
```

```python
nums.reverse() #also descending order
nums
```

####Operations

```python
print(len(names))      #length of list
```

```python
nums = [10, 5, 20]

print(len(nums))   # 3
print(min(nums))   # 5
print(max(nums))   # 20
print(sum(nums))   # 35
```

####Copying Lists

❌ Wrong way (reference copy)

```python
a = [1, 2]
b = a
print(a)

#Change b → a also changes
```

✅ Correct ways

```python
a = [5, 7]
b = a.copy()
b
```

```python
a = [91, 42]
b = a[:]
b
```

####**Lists + Loops**

###**Lists + if**

```python
scores = [45, 78, 90, 32]

for s in scores:
    if s >= 50:
        print("Pass")
    else:
        print("Fail")
```

1️⃣ Space-separated numbers (most common)

User types:
1 2 3 4 5

lst = list(map(int, input().split()))
print(lst)


📌 Output:

[1, 2, 3, 4, 5]


What happens

input() → "1 2 3 4 5"

.split() → ["1", "2", "3", "4", "5"]

map(int, ...) → convert to integers

list() → make a list

2️⃣ Space-separated strings

User types:
apple banana orange

lst = input().split()
print(lst)


📌 Output:

['apple', 'banana', 'orange']


(No map needed because they are already strings.)

3️⃣ Numbers entered one by one (using loop)
